```python
def corpus_def(urns):
    metadata = mm.metadata(urns)
    urns = [u for u in metadata if metadata[u] != []]
    return nb.frame({u:metadata[u] for u in urns})
```

```python
def evaluate(temaer, corpus):
    theme_eval = nb.wordbag_eval(temaer,list(corpus.index))
    # legger til metadata
    result = theme_eval.join(titlestmnt(corpus))
    return result
```

```python
def plot_together(A, B, aname = "A", bname = "B", rot = 60):
    df = pd.DataFrame()
    df[aname] = A
    df[bname] = B
    plotfig(df.sort_values(by=aname, ascending=False).plot(kind='bar', rot=60, figsize=(15,5), title="Comparison mean conditional"))
    return
```

# Fordeling av nasjoner

```python
def order_cats(df,  columns = 14):
    if isinstance(columns, int):
        # number of columns used
        res = {x:len(df[df[x] > 0]) for x in df.columns[:columns]}
    else:
        # actual columns
        res = {x:len(df[df[x] > 0]) for x in df[columns]}
    return nb.frame_sort(nb.frame(res))
           
```

```python
def nations_sample(samplesize):
    df_nations = pd.concat([
    order_cats(adult_eval.sample(samplesize, axis = 0), columns = nations_),
    order_cats(children_eval.sample(samplesize, axis = 0), columns = nations_)
], axis = 1)
    df_nations.columns = ['Adult', 'Children']
    return df_nations

def features_sample(samplesize):
    df_features = pd.concat([
    order_cats(adult_eval.sample(samplesize, axis = 0), columns = features_),
    order_cats(children_eval.sample(samplesize, axis = 0), columns = features_)
], axis = 1)
    df_features.columns = ['Adult', 'Children']
    return df_features
```

```python
def cross(df, cols = None):
    if isinstance(cols, (list, object)):
        res = df[cols].transpose().dot(df[cols])
    else:
        res = df.transpose().dot(df)
    return res
```

```python
def cosine_sims(df):
    """ similarity takes rows, so there must be transpose()"""
    cosines = pd.DataFrame(cosine_similarity(df.transpose()))
    cosines.columns = df.columns
    cosines.index = df.columns
    return cosines
     
```

```python
def cosine_dists(df):
    """ similarity takes rows, so there must be transpose()"""
    cosines = pd.DataFrame(cosine_distances(df.transpose()))
    cosines.columns = df.columns
    cosines.index = df.columns
    return cosines
     
```
