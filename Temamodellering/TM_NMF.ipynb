### Angi tekstene, hvor de ligger og bygg korpuset

List opp tekstene som skal inn, med sti relativ til notebooken. Bøkene listes her, fra en og oppover.


```python
txt = ["korpus/ad3.txt", "korpus/constance.txt", "korpus/sand-indiana.txt"]
```

```python
# norsk tekst

corpus=[]
for filename in txt:
    text = open(filename, encoding='utf-8').read()
    for paragraph in text.split('\n'):
        corpus.append(tokenize(paragraph))
```

```python
# velg noen avsnitt for å få et inntrykk - 5 eller 10 om gangen.

for p in corpus[100:105]:
    print(' '.join(p))
```

Bygg en frekvensliste

```python
freqs = tm.corpusfreq(corpus)
```

```python
freqs.most_common(20)
```

### Lag  en stoppordliste

Ordlista kan endres ved å legge til og trekke fra. Viser nedenfor hvordan det gjøres.

Ett parameter som er enkelt å endre er tallet i **freqs.most_common()**, bare endre det for å få andre stoppordlister. Ord som er tatt ut kan legges inn igjen. Se et par celler nedenfor.

```python
korpus_stopp = dict({x[0]:freqs[x[0]] for x in freqs.most_common(100)})
```

**Sjekk ut hva som fjernes...**

```python
print(korpus_stopp.keys())
```

Kjør så en modellering av tema. Bare endre på tallene for å se hvordan resultatet blir

```python
tm.modeller_set(corpus, tema=20, chunk_size=100, stop_words = korpus_stopp, topic_size=15)
```
